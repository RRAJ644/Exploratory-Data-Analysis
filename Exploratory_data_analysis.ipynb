```python
# Outlier treatment

# 1. 3 Sigma Technique
data = pd.read_csv('/raw_sales.csv')
data.head()
```

```python
# Function to detect outlier on 1d datasets

def findAnomalies(data):
  anomalies = []
  randomDataStd = statistics.stdev(data)
  randomDataMean = statistics.mean(data)

  anomalyCutOff = randomDataStd * 3

  lowerLimit = randomDataMean - anomalyCutOff
  upperLimit = randomDataMean + anomalyCutOff

  for outlier in data:
    if outlier > upperLimit or outlier < lowerLimit:
      anomalies.append(outlier)
  return anomalies
```
